```python
import matplotlib.pyplot as plt
import numpy as np

# Example data
x = np.linspace(-2, 2, 100)
y = np.exp(-x**2)
y2 = np.exp(-(x-1)**2)

plt.figure(figsize=(8, 4))
plt.plot(x, y, label='Normal Distribution')
plt.fill_between(x, y, color='orange', alpha=0.5)
plt.plot(x, y2, label='Shifted Distribution')
plt.title('Example Plot')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

x = np.linspace(0, 10, 100)
y = np.linspace(0, 1, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y * np.pi)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis')
fig.colorbar(surf, shrink=0.5, aspect=5)  # Add a color bar which maps values to colors.
plt.xlabel('Input')
plt.ylabel('Primary Membership')
ax.set_zlabel('Secondary Membership')
plt.show()
```

```python
from graphviz import Digraph

# Create a new directed graph
dot = Digraph(comment='The Round Table')

# Set global node attributes
dot.attr(rankdir='LR', shape='ellipse')

# Adding nodes with specific styles
dot.node('Input', 'X_t')
dot.node('Output', 'Y')
dot.node('Hidden', 'Z_t,k', shape='box', style='filled', color='lightblue')

# Adding edges with labels
dot.edge('Input', 'Hidden', label='weights', fontsize='10')
dot.edge('Hidden', 'Output', label='activation', fontsize='10')

# Subgraph for Layer 1
with dot.subgraph(name='cluster_0') as c:
    c.attr(style='filled', color='lightgrey')
    c.node_attr.update(style='filled', color='white')
    c.edges([('a0', 'a1'), ('a1', 'a2'), ('a2', 'a3')])
    c.attr(label='Layer 1')

# Subgraph for Layer 2
with dot.subgraph(name='cluster_1') as c:
    c.node_attr.update(style='filled', color='blue')
    c.edges([('b0', 'b1'), ('b1', 'b2'), ('b2', 'b3')])
    c.attr(label='Layer 2')

# Additional customization
dot.edge('X_t', 'a0')
dot.edge('b3', 'Y')

# Render the output to a file (PDF, PNG, etc.)
dot.render('output_file_name', view=True)  # Set view=False if you don't want it to open automatically.

# To print the source code of the graph to the console
print(dot.source)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Example data: generate synthetic data for demonstration
np.random.seed(0)
iterations = np.arange(0, 201, 1)
lines = ['API', 'REG', 'DLL+REG', 'DLL', 'API+REG', 'API+DLL+REG', 'API+DLL']
data = {line: np.cumsum(np.random.randn(201) * 0.05 - line.count('+') * 0.02 - i * 0.01) for i, line in enumerate(lines)}

plt.figure(figsize=(12, 8))

for i, (key, values) in enumerate(data.items()):
    mean_line = values
    sigma = np.std(values) / 4  # Adjust the standard deviation scaling as needed
    plt.plot(iterations, mean_line, label=key, lw=2)
    plt.fill_between(iterations, mean_line - sigma, mean_line + sigma, alpha=0.2)

plt.title('Mean & 0.5*Std.')
plt.xlabel('Iteration')
plt.ylabel('Best Cost')
plt.legend()
plt.grid(True)
plt.ylim(-0.2, 1.0)  # Set the limits for the y-axis
plt.xlim(0, 200)  # Set the limits for the x-axis
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the seed for reproducibility
np.random.seed(0)

# Generate synthetic data
iterations = np.arange(0, 201)
base = np.linspace(0, 1, iterations.size)**3  # Simulate a decreasing trend
lines = {
    'API': base + np.random.normal(0, 0.02, iterations.size),
    'REG': base + np.random.normal(0, 0.02, iterations.size) - 0.05,
    'DLL+REG': base + np.random.normal(0, 0.02, iterations.size) - 0.1,
    'DLL': base + np.random.normal(0, 0.02, iterations.size) - 0.15,
    'API+REG': base + np.random.normal(0, 0.02, iterations.size) - 0.2,
    'API+DLL+REG': base + np.random.normal(0, 0.02, iterations.size) - 0.25,
    'API+DLL': base + np.random.normal(0, 0.02, iterations.size) - 0.3,
}

# Create the plot
plt.figure(figsize=(12, 6))
colors = plt.cm.viridis(np.linspace(0, 1, len(lines)))  # Color map for lines

for (line, data), color in zip(lines.items(), colors):
    mean = np.mean(data)
    std = np.std(data)
    plt.plot(iterations, data, label=line, color=color)
    plt.fill_between(iterations, data - 0.5*std, data + 0.5*std, color=color, alpha=0.1)

plt.title('Mean & 0.5*Std.')
plt.xlabel('Iteration')
plt.ylabel('Best Cost')
plt.legend(title='Legend')
plt.grid(True)
plt.ylim(-0.2, 1.0)
plt.xlim(0, 200)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the seed for reproducibility
np.random.seed(0)

# Generate synthetic data for demonstration
iterations = np.arange(0, 201)
base = np.linspace(0, 1, iterations.size)**3  # Simulate a decreasing trend

# Adding a consistent decrement to simulate the separation between lines
lines = {
    'API': base + np.random.normal(0, 0.02, iterations.size),
    'REG': base + np.random.normal(0, 0.02, iterations.size) - 0.05,
    'DLL+REG': base + np.random.normal(0, 0.02, iterations.size) - 0.1,
    'DLL': base + np.random.normal(0, 0.02, iterations.size) - 0.15,
    'API+REG': base + np.random.normal(0, 0.02, iterations.size) - 0.2,
    'API+DLL+REG': base + np.random.normal(0, 0.02, iterations.size) - 0.25,
    'API+DLL': base + np.random.normal(0, 0.02, iterations.size) - 0.3,
}

plt.figure(figsize=(10, 6))
colors = ['purple', 'blue', 'green', 'red', 'orange', 'brown', 'pink']  # Specifying color for each line

for (line, data), color in zip(lines.items(), colors):
    mean = np.mean(data)
    std = np.std(data) / 2  # Adjust std deviation to better match the original image
    plt.plot(iterations, data, label=line, color=color)
    plt.fill_between(iterations, data - std, data + std, color=color, alpha=0.1)  # Lighter fill

plt.title('Mean & 0.5*Std.')
plt.xlabel('Iteration')
plt.ylabel('Best Cost')
plt.legend(loc='upper right', title='Legend')
plt.grid(True)
plt.ylim(-0.2, 1.0)
plt.xlim(0, 200)
plt.show()
```
